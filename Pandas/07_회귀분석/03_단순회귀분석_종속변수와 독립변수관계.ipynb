```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./file_new/auto-mpg_v2.csv")
```

```python
df
```

```python
# 지도학습 모델: 단순회귀분석(선형회귀모델,선형회귀예측모델..)
# - 종속변수가 있어요..
# - 직선의 방정식을 사용 : y = ax + b 
# - y: 종속변수  (정답)
# - x: 독립변수..
# - a: 기울기
# - b: y절편

# - 학습: fit
```

```python
# 사용할 변수들: mpg,cylinders,horsepower,weight
ndf = df[['mpg','cylinder','horsepower','weight']]
ndf
```

```python
#산점도 그래프 그리기..
# 종속변수:mpg, 독립변수 : weight 
ndf.plot(kind = "scatter",
        x="weight", 
        y="mpg", 
        color = 'g',
         s = 10, #사이즈의미
        figsize = (10,5))
sns.regplot(x='weight', y='mpg', data=ndf,
           line_kws={'color': 'red'}, scatter_kws={'color': 'grey'})

ndf.plot(kind = "scatter",
        x="weight", 
        y="mpg", 
        color = 'g',
         s = 10, #사이즈의미
        figsize = (10,5))

plt.show()
plt.close()
```

```python
#회귀선(추세선) 그리기
fig = plt.figure(figsize = (10,5))

# 1번째 그래프 공간 만들기
ax1 = fig.add_subplot(1, 2, 1)

# 2번째 그래프 공간 만들기
ax2 = fig.add_subplot(1, 2, 2)

# 첫번째 그래프
sns.regplot(x = "weight",
           y = "mpg",
            data = ndf,
           ax = ax1,
           color = "red",
            marker ="x",
           fit_reg = True)

sns.regplot(x = "weight",
           y = "mpg",
            data = ndf,
           ax = ax2,
            color = "blue",
           fit_reg = False)

plt.show()
```

```python
fig = plt.figure(figsize = (30, 30)) #사진크기지정
sns.jointplot(x = "weight", y = "mpg", data = ndf, kind='reg') #x,y지정 후 data=gen으로 지정 종류는 scatter
```
