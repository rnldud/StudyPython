```python
import pandas as pd
import numpy as np
```

```python
#데이터셋 불러오기
file_path = "./data/bicycle.csv"
df = pd.read_csv(file_path, encoding="euc-kr")  #euc-kr 우리나라만 가능한 데이터 타입 
```

```python
#전체 컬럼 및 타입확인 (결측치확인을위해)
df.info()
```

```python
#결측 데이터 확인
df.isnull()
df.notnull()
```

```python
#컬럼별 결측값이 있는 행의 갯수를 조회
df.isnull().sum(0)
```

```python
#행별 결측값이 있는 컬럼의 갯수 조회:sum()
df.isnull().sum(1)
```

```python
df.notnull().sum(0)
```

```python
#결측치가 있는 모든 행 삭제..
df.dropna(axis=0)
```

```python
#결측치가 있는 모든행 삭제
df_drp_allrow = df.dropna(axis = 0)
df_drp_allrow.head() #앞열만 확인
```

```python
#컬럼단위 제거 
df_drp_allcol = df.dropna(axis=1)
df_drp_allcol.head()
```

```python
#기준 col을 찾아 결측치가 있는 행삭제

df["대여거치대"].dropna(axis=0) #df.대여거치대.dropna(axis=0)가능 
```

```python
df["대여거치대"].notnull().sum()
```

```python
#여러개 컬럼을 기준으로 삭제하기
df[["대여소번호",
    "대여거치대", 
    "이용시간"]].dropna(axis=0).head()
```

```python
#결측치를 모두 0으로 채우기
df.fillna(0)
```

```python
#컬럼 이용거리 결측치값 0으로 변경
df["이용거리"].fillna(0)
#df.이용거리.fillna(0)
```

```python
#이용거리 컬럼 중 결측치 miising으로 수정
df["이용거리"].fillna("missing")
#df.이용거리.fillna("missing")
```

```python
#평균으로 대체값 사용하기 
#전체 데이터 중 결측치가 있으며
#전체 평균값으로 채워주시오
df.fillna(df.mean())
```

```python
#이용거리 결측치가 있을 경우 평균값으로 채워주시오
df["이용거리"].fillna(df["이용거리"].mean())
```
