# 중복 데이터 처리하기

### 데이터 가지고 오기

```python
#bicycl_dup.csv 읽어들이신 후 조회해주세요..
import pandas as pd
```

```python
#데이터셋 읽어들이기
file_path = "./data/bicycle_dup.csv"
df = pd.read_csv(file_path,encoding="euc-kr")

df
```

```python
# 중복데이터 찾기..
df.duplicated()
# 한개컬럼만 찾기
# duplicated에 keep을 지정안하면..
# 중복값중에 첫번째 값은 False, 
# 중복된 값중에 첫번째를 제외한것은 모두 True
# 제거할 때는 True항 만 제거 

df.duplicated(["이용거리"])

#df전체(전체컬럼에 전체 행을 의미)데이터중에
#"이용거리"라는 컬럼에 대해서만 중복체크 
```

```python
# 데이터 프레임에 이용거리 컬럼에 대해서 중복체크해주시오
df["이용거리"].duplicated()
```

```python
# 두개 컬럼을 기준으로 중복체크하기
# 이용거리, 나이
df.duplicated(["이용거리","나이"])

df[["이용거리","나이"]].duplicated()
```

```python
#중복이 있는 데이터 개수 구하기 
len(df[df.duplicated(["이용거리","나이"])])
```

```python
len(df.loc[df.duplicated(['이용거리','나이'])])
```

```python
# 중복데이터 -keep = "first"
# keep 생략하면 기본: keep = "first" 있다
# 가장많이 사용하는 중복체크 방식...

df.duplicated(['이용거리'],keep = "first")
```

```python
# keep ="last"
df.duplicated(['이용거리'],keep = "last")
```

```python
# keep = False
# 중복이 하나라도 있으면 True 
# 중복이 있는 모든데이터 True로 반환시키자!
df_dup = df.duplicated(["이용거리","나이"], keep = False)
```

```python
#이용거리,나이, 두개 컬럼을 기준으로 중복되는 모든 데이터를 조회해주시오 
len(df[df_dup])
```

```python
# 첫번째 중복값만 남기고 나머지 제거하기 
# 실행 결과물은 제거된 후에 정상 데이터만 조회시킴 

#삭제시 duplicates 

df.drop_duplicates(['이용거리'],keep = 'first')
```

```python
# 마지막 중복값들만 남기고, 제거하기
df.drop_duplicates(["이용거리"], keep = "last")
```

```python
# keep = "first" 와 "last"의 결과값은 같아야함 
```

```python
# 중복된 데이터 전체 삭제하기
df.drop_duplicates(["이용거리"], keep = False)
```

```python
df_dup = df.drop_duplicates(["이용거리"], keep = False)
```

```python
# 완전히 깨끗한 데이터
df_dup
```

```python
#중복 검증
df_dup[df_dup.duplicated(["이용거리"], keep = False)]
```
