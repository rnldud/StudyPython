## 최대값/최소값 계산하여 제거하기

```python
# 넘파이 라이브러리 불러들이기
import numpy as np
```

```python
# q1, q3 값 확인하기 
# percentile(): 25% 와 75% 시점의 데이터 값을 알려줌 
# 두개의값을 받을 때 변수 선언 방식 (q1,q3)
q1,q3 = np.percentile(df_drop_allrow["나이"],[25, 75]) 
print(q1) #최소값
print(q3) #최대값 
```

```python
# iqr 값 계산하기
iqr = q3-q1

print(iqr)
```

```python
#최대값 최소값 알아내기
#최대값 / 최소값 계산하기
#최대값 q3 + 1.5(q3-q1)
#최소값 q1 - 1.5(q3-q1)
```

```python
#최대값 계산하기
upper_bound = q3 + (iqr * 1.5)
upper_bound
```

```python
#최소값 계산하기
lower_bound = q1 - (iqr * 1.5)
lower_bound
```

```python
# 나이 데이터에서 이상치값 조회하기..
df_drop_allrow[(df_drop_allrow["나이"]> upper_bound) | (df_drop_allrow["나이"]< lower_bound)]

#갯수확인 원하면,
#len(df_drop_allrow[(df_drop_allrow["나이"]> 61.5) | (df_drop_allrow["나이"]< 1.5)])
```

```python
# 매개변수가 없는 함수로 만들어보자 
# 반복해서 사용되거나, 자주 사용되는 목적으로 만들어짐
# 그렇기에, 함수는 범용으로 사용할 수 있으면 좋다
# 기능을 수행하는 목적이 가장 큼
def outliers_iqr(): #함수를 만들어 시작하겠도다 
    #q1, q3 값 확인하기
    #percentile(): 25%와 75% 시점의 데이터 값을 알려줌(return)
    q1,q3 = np.percentile(df_drop_allrow["나이"],[25, 75]) 
    print(q1) 
    print(q3) 
    
    #iqr 값 계산
    iqr = q3-q1   
    print(iqr)
    
    #최대값 계산
    upper_bound = q3 + (iqr * 1.5)
    print(upper_bound)
    
    #최소값 계산
    lower_bound = q1 - (iqr * 1.5)
    print(lower_bound)
    
    #나이 데이터에서 이상치 값 조회하기 
    df_temp = df_drop_allrow[(df_drop_allrow["나이"]> upper_bound) | 
                             (df_drop_allrow["나이"]< lower_bound)]
    print(len(df_temp))
    print(df_temp)
```

```python
# 데이터를 넘겨주는 대로 알아서 이상치 값을 출력하는 기능만들기 
# 매개변수가 있는 함수 만들기
import numpy as np

def outliers_iqr2(data): 
    #q1, q3 값 확인하기
    #percentile(): 25%와 75% 시점의 데이터 값을 알려줌(return)
    q1,q3 = np.percentile(data,[25, 75]) 
    print(q1) 
    print(q3) 
    
    #iqr 값 계산
    iqr = q3-q1   
    print(iqr)
    
    #최대값 계산
    upper_bound = q3 + (iqr * 1.5)
    print(upper_bound)
    
    #최소값 계산
    lower_bound = q1 - (iqr * 1.5)
    print(lower_bound)
   
    #나이 데이터에서 이상치 값 조회하기 
    df_temp = df_drop_allrow[(data> upper_bound) | 
                             (data< lower_bound)] #전체 조회위해 data로 변경 
    print(len(df_temp))
    print(df_temp)
```

```python
# 매개변수가 2개인 함수 만들기 
def outliers_iqr3(dframe, data): 
    #q1, q3 값 확인하기
    #percentile(): 25%와 75% 시점의 데이터 값을 알려줌(return)
    q1,q3 = np.percentile(data,[25, 75]) 
    print(q1) 
    print(q3) 
    
    #iqr 값 계산
    iqr = q3-q1   
    print(iqr)
    
    #최대값 계산
    upper_bound = q3 + (iqr * 1.5)
    print(upper_bound)
    
    #최소값 계산
    lower_bound = q1 - (iqr * 1.5)
    print(lower_bound)
    
    #나이 데이터에서 이상치 값 조회하기 
    df_temp = dframe[(data> upper_bound) | 
                     (data< lower_bound)] #전체 조회위해 data로 변경 
    print(len(df_temp))
    print(df_temp)
```

```python
# 함수 호출하기
# 이제 이 함수만 불러오면, 이상치를 확인 할 수 있음
outliers_iqr3(df_drop_allrow, df_drop_allrow["나이"])
```

### 이상치 제거하기 

```python
# 이상치가 아닌 값만 추출 
# 아닌값 따로 저장하기 

age_non_outlier_index = []

#이상치 인덱스가 아닌 정상 인덱스만 추출하기
for idx in df_drop_allrow.index :   
    if idx not in age_outlier_index[0] : #array 형태인 경우 [0]번째 지정해주어야함 
        age_non_outlier_index.append(idx)

print(len(age_non_outlier_index))
age_non_outlier_index        
```

```python
# 전체 데이터 프레임에서 정상 데이터만 조회해주세요 
# 이상치가아닌 정상데이터 
df_new = df_drop_allrow.loc[age_non_outlier_index] 

df_new
```
