## 2. 재구조화

```python
#cut()
df1 = pd.cut(df["math"],3)
df1
```

```python
df2 = df["math"].groupby(df1)
```

```python
df2.head(1)
```

```python
# 기초통계 (구간별통계) : agg() 함수사용
df3 = df2.agg(["count","mean","std","min","max"])
df3
```

```python
#qcut()
pd.qcut(df["math"],3)
```

```python
#labels=False : 범주의 번호를 표시하는 것
df4 = pd.qcut(df["math"], 3,labels=False)
```

```python
#위 구간과 학인해보기!
df4
```

```python
#arange를 이용해 거꾸로 도출!
import numpy as np

df5 = pd.qcut(df["math"], 3,labels = np.arange(3,0,-1))
```

```python
#첫번째값: 시작값
#두번째값: 끝값 -1
#세번째값: 증가값 or 감소값 
for i in range(0,3,1) : #0부터 시작해서 1씩증가
    print(i)

print("------")

for j in range(3,0,-1) : #3부터 시작해서 1씩감소 
    print(j)
```

```python
#첫번째값: 시작값
#두번째값: 끝값 -1
#세번째값: 증가값 or 감소값 

for i in np.arange(0,3,1) :
    print(i)
```

```python
df5
```

```python
df6 = df["math"].groupby(df5)
```

```python
df6.head(1)
```

```python
#기초통계 조회
df7 = df6.agg(["count","mean","std","min","max"])
df7

#한명밖에없어서 중앙값 nan
```

### 3. 원-핫 인코딩

```python
# 기계가 이해할 수 있는 형태로 데이터 변환. 
# 범주형 데이터를 원-핫인코딩형태로 변환
# 범주형 데이터 0 or 1로 변환 
```

```python
# df의 데이터타입을 확인해보시오! #dtypes
# object가 일반적으로 범주형 데이터 
df.dtypes
```

```python
# 범주형 컬럼에 대해서만 가능
# 범주형이 아닌 컬럼에 대해선,,범주형으로 만들면 가능 
# get_dummies() : 숫자값을 제외한 object 데이터 타입만 가능
# 해당 one_hot 인코딩 되는 원본 컬럼은 없어지고,
# 해당 컬럼명_범주명으로 컬럼이 만들어진다 
df_onehot = pd.get_dummies(df) #원핫인코딩은 get_dummies함수만 사용
```
