```python
import os, glob
filter(os.path.isdir, glob.glob('./*'))
dataList = glob.glob('./*')

folderList = []
for x in dataList:
    # print(x), # print(x[2:])           
    # print('-'*20)
    # print(x[2:])
    if len(x[2:].split('.')) == 2:   #print(x) = .\eee 일 경우 .을기준으로 split 그래서 == 2가되는 것 
                                 #print(x[2:]) = eee일 경우 .을기준으로 split 했을 경우 == 1 가능/ 2(확장자file)    
        folderList.append(x)
folderList
```

```python
import os, glob
filter(os.path.isdir, glob.glob('./*')) # . = 현재위치/* = sql의 all과 같은 의미/isdir = 디렉토리 
dataList = glob.glob('./*') #카멜(낙타)표기법_보기쉽게하기위해

#if x.split('.')[1] == 'jpg': 참고확장자에 대한정보가 jpg라면
folderList=[]
for x in dataList:   #객체 할당 
     if '.' not in x[2:]: # 맨앞'.'은 현재 위치를 설명하는 것, 그래서 앞쪽'.'이아닌 확장자 '.' 도출을 원함
         folderList.append(x)
folderList

# # for x in range(len(dataList)):   #인덱스 할당 같은 방식이지만, 이방법은 위 방법보다 느림.
# #     print(dataList[0])
```
